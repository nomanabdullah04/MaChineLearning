# **Feature Engineering**

```python
##Getting all different types of feature
num_feature=[feature for feature in df.columns if df[feature].dtype!='O']
cat_feature=[feature for feature in df.columns if df[feature].dtype=='O']
discrite_feature=[feature for feature in num_feature if len(df[feature].unique())<25]
continous_feature=[feature for feature in num_feature if feature not in discrite_feature]
print(len(num_feature))
print(cat_feature)
print(discrite_feature)
print(continous_feature)
```

```python
##Remove Unnecessary Columns
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.drop(['car_name'],axis=1,inplace=True)
df.drop(['brand'],axis=1,inplace=True)
df.head()
```

```python
#Define idpendent and dependent feature
x=df.drop(['selling_price'],axis=1)
y=df['selling_price']
```

# ***Feature Encoding and Scaling***

```python
df['model'].value_counts()
```

```python
#Create column Transformer with 3 types of transformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler

# Define num_feature and cat_feature based on the state of x after label encoding 'model'
# and before conversion to csr_matrix.
# 'Unnamed: 0', 'model', 'vehicle_age', 'km_driven', 'mileage', 'engine', 'max_power', 'seats' were numeric.
# 'car_name', 'brand', 'seller_type', 'fuel_type', 'transmission_type' were categorical.
num_feature = ['model', 'vehicle_age', 'km_driven', 'mileage', 'engine', 'max_power', 'seats']
cat_feature = ['seller_type', 'fuel_type', 'transmission_type']

print(num_feature)
print(cat_feature)
len(df['seller_type'].unique())
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
#Label encoding to the feature model
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
x['model'] = label_encoder.fit_transform(x['model'])
```

```python
numerical_transformer=StandardScaler()
categorical_transformer=OneHotEncoder(drop='first')
preprocessor=ColumnTransformer(
    [
     ('OneHotEncoder',categorical_transformer,cat_feature),
     ('StandardScaler',numerical_transformer,num_feature)
    ],remainder='passthrough'
)
label_encoder_columns=['model']
```

```python
x=preprocessor.fit_transform(x)
```

```python
print(pd.DataFrame(x))
```

# **Train and Test splits the models**

As you can see, the `original_car_name` column now represents the car names in a single column again. You can apply a similar process to other one-hot encoded features if needed.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train.shape,x_test.shape
```

# **Model Traning and Model Selection**

```python
#Create a function to Evalute the model
def evaluate_model(true,predicted):
  mae=mean_absolute_error(true,predicted)
  mse=mean_squared_error(true,predicted)
  rmse=np.sqrt(mean_squared_error(true,predicted))
  r2=r2_score(true,predicted)
  return mae,mse,rmse,r2
```

```python
#Begining model traning
import numpy as np
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor,BaggingRegressor
from sklearn.svm import SVR

#Create a function to Evalute the model
def evaluate_model(true,predicted):
  mae=mean_absolute_error(true,predicted)
  mse=mean_squared_error(true,predicted)
  rmse=np.sqrt(mean_squared_error(true,predicted))
  r2=r2_score(true,predicted)
  return mae,mse,rmse,r2

models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'KNeighborsRegressor':KNeighborsRegressor(),
    'DecisionTreeRegressor':DecisionTreeRegressor(),
    'RandomForestRegressor':RandomForestRegressor(),
    'GradientBoostingRegressor':GradientBoostingRegressor(),
    'AdaBoostRegressor':AdaBoostRegressor(),
    'ExtraTreesRegressor':ExtraTreesRegressor(),
    'BaggingRegressor':BaggingRegressor(),
    'SVR':SVR()
}
for i in range(len(list(models))):
  model=list(models.values())[i]
  model.fit(x_train,y_train)
  y_train_pred=model.predict(x_train)
  y_test_pred=model.predict(x_test)
  model_train_mae,model_train_mse,model_train_rmse,model_train_r2=evaluate_model(y_train,y_train_pred)
  model_test_mae,model_test_mse,model_test_rmse,model_test_r2=evaluate_model(y_test,y_test_pred)
  print(list(models.keys())[i])
  print('Model performance for Training set')
  print("- Root Mean Squared Error: {:.4f}".format(model_train_rmse))
  print("- Mean Squared Error: {:.4f}".format(model_train_mse))
  print("- Mean Absolute Error: {:.4f}".format(model_train_mae))
  print("- R2 Score: {:.4f}".format(model_train_r2))
  print('----------------------------------')
  print('Model performance for Test set')
  print("- Root Mean Squared Error: {:.4f}".format(model_test_rmse))
  print("- Mean Squared Error: {:.4f}".format(model_test_mse))
  print("- Mean Absolute Error: {:.4f}".format(model_test_mae))
  print("- R2 Score: {:.4f}".format(model_test_r2))
  print('='*35)
  print('\n')
```

# *By Using Hyperperameter tuning*

```python
#Initialize few parameter for hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
knn_param={
    'n_neighbors':np.arange(1,30),
    'weights':['uniform','distance'],
    'algorithm':['auto','ball_tree','kd_tree','brute']
}

rf_param={
    'n_estimators':[100,200,500,1000],
    'max_depth':[3,5,10,20],
    'min_samples_split':[2,5,10,15],
    'min_samples_leaf':[1,2,5,10]
}
```

```python
#Model list for hyperparamaeter tuning
randomcv_model={
    'KNeighborsRegressor':KNeighborsRegressor(),
    'RandomForestRegressor':RandomForestRegressor(),
}
```

```python
model_param={}

for name,model,param in zip(randomcv_model.keys(),randomcv_model.values(),[knn_param,rf_param]):
  random=RandomizedSearchCV(estimator=model,param_distributions=param,n_iter=100,cv=5,verbose=2,n_jobs=-1,random_state=42)
  random.fit(x_train,y_train)
  model_param[name]=random.best_params_

  for model_name in model_param:
    print(f"{model_name} best params: {model_param[model_name]}")
    print('='*35)
```

```python
#Retraning the models with best parameters
models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'KNeighborsRegressor':KNeighborsRegressor(algorithm='auto',n_neighbors=27,weights='distance'),
    'DecisionTreeRegressor':DecisionTreeRegressor(),
    'RandomForestRegressor':RandomForestRegressor(max_depth=20,min_samples_leaf=1,min_samples_split=10,n_estimators=1000),
    'GradientBoostingRegressor':GradientBoostingRegressor(),
    'AdaBoostRegressor':AdaBoostRegressor(),
    'ExtraTreesRegressor':ExtraTreesRegressor(),
    'BaggingRegressor':BaggingRegressor(),
    'SVR':SVR()
}
```

```python
for i in range(len(list(models))):
  model=list(models.values())[i]
  model.fit(x_train,y_train)
  y_train_pred=model.predict(x_train)
  y_test_pred=model.predict(x_test)
  model_train_mae,model_train_mse,model_train_rmse,model_train_r2=evaluate_model(y_train,y_train_pred)
  model_test_mae,model_test_mse,model_test_rmse,model_test_r2=evaluate_model(y_test,y_test_pred)
  print(list(models.keys())[i])
  print('Model performance for Training set')
  print("- Root Mean Squared Error: {:.4f}".format(model_train_rmse))
  print("- Mean Squared Error: {:.4f}".format(model_train_mse))
  print("- Mean Absolute Error: {:.4f}".format(model_train_mae))
  print("- R2 Score: {:.4f}".format(model_train_r2))
  print('----------------------------------')
  print('Model performance for Test set')
  print("- Root Mean Squared Error: {:.4f}".format(model_test_rmse))
  print("- Mean Squared Error: {:.4f}".format(model_test_mse))
  print("- Mean Absolute Error: {:.4f}".format(model_test_mae))
  print("- R2 Score: {:.4f}".format(model_test_r2))
  print('='*35)
  print('\n')
```
