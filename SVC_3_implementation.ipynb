# Data Preprocessing: Categorical Encoding

This section covers how to transform text-based categorical columns into numerical values so that machine learning algorithms (like SVMs) can process them.



## 1. Core Concepts

### Label Encoding
Converts each unique category into a standalone integer ($0, 1, 2, \dots$).
* **Best Use Case:** Ordinal data (categories with a natural rank/order, like *Low, Medium, High*).
* **Limitation:** Models may mistakenly assume a mathematical hierarchy ($2 > 0$) for unordered categories.

### One-Hot Encoding
Creates a brand-new binary column ($0$ or $1$) for every unique category in the dataset.
* **Best Use Case:** Nominal data (categories without any internal order, like *Red, Green, Blue* or *City Names*).
* **Limitation:** Increases dataset width significantly if a column contains many unique categories.


## 2. Quick Summary Table


| Data Type | Example | Recommended Encoding | Reason |
| :--- | :--- | :--- | :--- |
| **Ordinal** | Small, Medium, Large | **Label Encoding** | Preserves the natural ranking. |
| **Nominal** | Paris, Tokyo, New York | **One-Hot Encoding** | Treats all categories equally. |


## 3. Visual Example of Transformations

### Label Encoding Example
* `High School` $\rightarrow$ `0`
* `Bachelor's` $\rightarrow$ `1`
* `Master's` $\rightarrow$ `2`

### One-Hot Encoding Example


| Original Feature (City) | City_New York | City_Paris | City_Tokyo |
| :--- | :---: | :---: | :---: |
| **New York** | 1 | 0 | 0 |
| **Paris** | 0 | 1 | 0 |
| **Tokyo** | 0 | 0 | 1 |




```python
#define indepndent and dependent feature
X = df.drop("total_bill", axis=1)
y = df["total_bill"]
```

```python
X
```

```python
#tran and split dataset

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.head()
```

```python
#Feature Endoding(label and onehot encoding)
from sklearn.preprocessing import LabelEncoder
le1= LabelEncoder()
le2= LabelEncoder()
le3= LabelEncoder()
X_train['sex']=le1.fit_transform(X_train['sex'])
X_train['smoker']=le2.fit_transform(X_train['smoker'])
X_train['time']=le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
#now for X_test data
X_test['sex']=le1.transform(X_test['sex'])
X_test['smoker']=le2.transform(X_test['smoker'])
X_test['time']=le3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
#Onehot encoding ------ columnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ohe= OneHotEncoder(sparse_output=False,drop='first') # Re-initialize with correct sparse_output
ct= ColumnTransformer(transformers=[('onehot',ohe,[3])],remainder='passthrough')
X_train=ct.fit_transform(X_train)
X_test=ct.transform(X_test)
```

```python
import pandas as pd
print(pd.DataFrame(X_train))
```

```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
print(pd.DataFrame(X_train))
```

```python
#Model traning
from sklearn.svm import SVR
svr= SVR(kernel='rbf')
svr.fit(X_train,y_train)
y_pred= svr.predict(X_test)
```

```python
#Print the accuracy and confused matrix
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("R2 Score: ",r2_score(y_test,y_pred))
print("MAE: ",mean_absolute_error(y_test,y_pred))
print("MSE: ",mean_squared_error(y_test,y_pred))
```

# **By using gridsearch CV**

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}
```

```python
grid_search = GridSearchCV(estimator=svr, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
```

```python
y_pred = grid_search.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```
