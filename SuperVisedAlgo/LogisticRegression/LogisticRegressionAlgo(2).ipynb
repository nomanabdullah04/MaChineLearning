# **Our Concept: Randomised Search CV**

# Hyperparameter Tuning: Why We Use RandomizedSearchCV

When optimizing a machine learning model, we must find the best configuration parameters (called **hyperparameters**), such as the regularization strength \(C\) or penalty type in Logistic Regression. The two main ways to automate this search are **GridSearchCV** and **RandomizedSearchCV**.


## 1. The Core Problem with GridSearchCV (The Exponential Trap)
GridSearchCV performs an exhaustive search. It acts like a brute-force algorithm that checks every single possible combination you provide in a grid.

###  The Grid Search Calculation
If you have 4 hyperparameters, and you want to test 10 possible values for each using 5-fold cross-validation (\(k=5\)):

\[\text{Total Training Runs} = (10 \times 10 \times 10 \times 10) \times 5 = 50,000 \text{ models}\]

If training a single model takes just 2 seconds, your Colab notebook will freeze or run for **nearly 28 hours** just to finish tuning.

## 2. How RandomizedSearchCV Solves This
Instead of checking every single intersection point on the grid, **RandomizedSearchCV** randomly selects a fixed number of combinations (controlled by the parameter `n_iter`) from a statistical distribution.

### Brute Force Grid vs. Randomized Allocation Diagram

```text
       GRID SEARCH (Exhaustive)              RANDOMIZED SEARCH (Budgeted)
       
       Param 2                               Param 2
          ▲                                     ▲
          │  ●   ●   ●   ●   ●                  │      ●          
          │  ●   ●   ●   ●   ●                  │          ●   ●  
          │  ●   ●   ●   ●   ●                  │  ●              
          │  ●   ●   ●   ●   ●                  │              ●  
          └───────────────────►                 └───────────────────►
                             Param 1                               Param 1
       (Tests all 20 intersections)          (Tests only 5 random points)
```


## 3. Why It Is Mathematically Better
1. **Budget Control:** You explicitly dictate the computing budget via `n_iter`. If you set `n_iter=20`, it will only pick 20 combinations, no matter how massive your parameter pool is.
2. **The "Important vs. Unimportant" Parameter Law:** In most models, only 1 or 2 hyperparameters truly impact performance, while others change very little.
   * **Grid Search** wastes dozens of runs repeating tests on the unimportant parameter.
   * **Randomized Search** explores distinct values across the important parameter space on every single iteration, discovering the optimal setup much faster.


## 4. Scikit-Learn Python Implementation
Copy this code pattern to see how to execute a randomized search for a Logistic Regression classifier:

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

# 1. Define your parameter distribution space
param_distributions = {
    'C': uniform(loc=0.1, scale=10),       # Continuous random distribution between 0.1 and 10.1
    'penalty': ['l1', 'l2'],                # Categorical choices
    'solver': ['liblinear', 'saga']         # Compatible solvers
}

# 2. Initialize the model and randomized search engine
lr = LogisticRegression(max_iter=1000)
random_search = RandomizedSearchCV(
    estimator=lr,
    param_distributions=param_distributions,
    n_iter=15,                             # Explicit budget: test exactly 15 random combinations
    cv=5,                                  # 5-Fold cross-validation
    random_state=42,
    n_jobs=-1                              # Use all available CPU cores to speed it up
)

# 3. Execute the search
# random_search.fit(X_train, y_train)
print("RandomizedSearchCV setup complete.")
```
