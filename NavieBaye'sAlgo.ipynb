## 1. Conditional Probability & Joint Events
This section introduces the foundational concepts of dependent sequential events using a visual marble example.

*   **Initial State:** A box contains 5 marbles total: 3 orange ($o$) and 2 yellow ($y$).
*   **First Event:** Removing an orange marble.
    $$\text{P}(o) = \frac{3}{5}$$
*   **Second Event:** Removing a yellow marble after an orange one has already been removed. Because one marble was removed without replacement, the total pool drops to 4 marbles (2 orange, 2 yellow).
    $$\text{P}(y \mid o) = \frac{2}{4} = \frac{1}{2}$$

### Joint Probability Formula
To find the probability of both events happening in sequence (drawing an orange and then a yellow marble), we use the multiplication rule:
$$\text{P}(o \text{ and } y) = \text{P}(o) \times \text{P}(y \mid o)$$


## 2. Derivation of Bayes' Theorem
Bayes' Theorem is derived mathematically from the commutative property of joint probabilities.

1. **Symmetry of Joint Probability:** The probability of $A$ and $B$ occurring is identical to the probability of $B$ and $A$ occurring:
   $$\text{P}(A \text{ and } B) = \text{P}(B \text{ and } A)$$

2. **Expansion via Multiplication Rule:** Replacing both sides with their conditional probability equivalents:
   $$\text{P}(A) \times \text{P}(B \mid A) = \text{P}(B) \times \text{P}(A \mid B)$$

3. **Bayes' Theorem Formula:** Solving for $\text{P}(A \mid B)$ by dividing both sides by $\text{P}(B)$:
   $$\text{P}(A \mid B) = \frac{\text{P}(B \mid A) \times \text{P}(A)}{\text{P}(B)}$$


## 3. Transitioning to Naive Bayes Classification
We can shift from basic probability to Machine Learning Classification by using a tabular dataset model.

*   **Input Features ($\mathbf{X}$):** Independent variables represented as $x_1, x_2, x_3$.
*   **Dependent Variable ($y$):** The target categorical classification output (e.g., `Yes` or `No`).

### Applying Bayes' Theorem to a Dataset
To predict a class $y$ given a set of input features, the theorem is structured as:
$$\text{P}(y \mid (x_1, x_2, x_3)) = \frac{\text{P}(y) \times \text{P}((x_1, x_2, x_3) \mid y)}{\text{P}(x_1, x_2, x_3)}$$

### The "Naive" Assumption
The core assumption of a Naive Bayes classifier is that **all input features are conditionally independent** given the class label. This allows us to break down the joint conditional probability in the numerator into a product of individual probabilities:

*   **For Class Yes:**
    $$\text{P}(\text{Yes} \mid (x_1, x_2, x_3)) = \frac{\text{P}(\text{Yes}) \times \text{P}(x_1 \mid \text{Yes}) \times \text{P}(x_2 \mid \text{Yes}) \times \text{P}(x_3 \mid \text{Yes})}{\text{P}(x_1) \times \text{P}(x_2) \times \text{P}(x_3)}$$

*   **For Class No:**
    $$\text{P}(\text{No} \mid (x_1, x_2, x_3)) = \frac{\text{P}(\text{No}) \times \text{P}(x_1 \mid \text{No}) \times \text{P}(x_2 \mid \text{No}) \times \text{P}(x_3 \mid \text{No})}{\text{P}(x_1) \times \text{P}(x_2) \times \text{P}(x_3)}$$


## 4. Practical Dataset Application (The "Play Tennis" Problem)
We use the classic "Play Tennis" dataset ($14$ total samples) to construct frequency and likelihood tables.

### Global Prior Probabilities
*   $\text{P}(\text{Yes}) = \frac{9}{14}$
*   $\text{P}(\text{No}) = \frac{5}{14}$

### Feature Likelihood Tables
Individual features are broken down against the target variable:
*   **Outlook:** Splits `Sunny`, `Overcast`, and `Rain` between Yes/No.
    *   $\text{P}(\text{Sunny} \mid \text{Yes}) = \frac{2}{9}$
    *   $\text{P}(\text{Sunny} \mid \text{No}) = \frac{3}{5}$
*   **Temperature:** Splits `Hot`, `Mild`, and `Cool`.
    *   $\text{P}(\text{Hot} \mid \text{Yes}) = \frac{2}{9}$
    *   $\text{P}(\text{Hot} \mid \text{No}) = \frac{2}{5}$

**Objective:** Predict the target outcome (`PLAY (Y/N)`) for a new data point where $\text{Outlook} = \text{Sunny}$ and $\text{Temperature} = \text{Hot}$.


## 5. Step-by-Step Calculation & Normalization
*Note: Because the denominator $\text{P}(\text{Sunny}) \times \text{P}(\text{Hot})$ is identical for both classes, it can be dropped during comparison.*

### Step 5.1: Calculate Raw Numerator for Class Yes
$$\text{P}(\text{Yes} \mid \text{Sunny}, \text{Hot}) \propto \text{P}(\text{Yes}) \times \text{P}(\text{Sunny} \mid \text{Yes}) \times \text{P}(\text{Hot} \mid \text{Yes})$$
$$\text{Substituting values:} \quad \frac{9}{14} \times \frac{2}{9} \times \frac{2}{9} = \frac{2}{63} \approx 0.031$$

### Step 5.2: Calculate Raw Numerator for Class No
$$\text{P}(\text{No} \mid \text{Sunny}, \text{Hot}) \propto \text{P}(\text{No}) \times \text{P}(\text{Sunny} \mid \text{No}) \times \text{P}(\text{Hot} \mid \text{No})$$
$$\text{Substituting values:} \quad \frac{5}{14} \times \frac{3}{5} \times \frac{2}{5} = \frac{3}{35} \approx 0.085$$

### Step 5.3: Normalization to Percentages
To find the actual probabilities, divide each raw score by their total sum ($0.031 + 0.085 = 0.116$):

*   **Final Probability for Yes:**
    $$\frac{0.031}{0.031 + 0.085} = \frac{0.031}{0.116} \approx 0.27 = 27\%$$

*   **Final Probability for No:**
    $$\frac{0.085}{0.031 + 0.085} = \frac{0.085}{0.116} \approx 0.73 = 73\%$$

### Conclusion
Since $73\% > 27\%$, the final classification outcome for a day that is **Sunny and Hot** is **No** (Do not play tennis).


# ***Advanced Naive Bayes: Variants and Specialized Data Types***

---




## 1. Overview of Naive Bayes Variants
The choice of a Naive Bayes variant depends entirely on the mathematical distribution and data type of the input features. The three primary algorithmic variants are:

*   **Bernoulli Naive Bayes:** Optimized for binary data.
*   **Multinomial Naive Bayes:** Optimized for discrete count data.
*   **Gaussian Naive Bayes:** Optimized for continuous numerical data.


## 2. Multinomial Naive Bayes & Text Processing
This variant is used when the input features consist of discrete frequency counts (e.g., word frequencies in a document).

*   **Core Use Case:** Text classification.
*   **Classic Problem:** Email Spam Filtering (`Spam` vs. `Ham`).
    *   *Spam Input Example:* "You have Million $$ lottery"
    *   *Ham Input Example:* "Krish you have done good job"

### The NLP Feature Extraction Pipeline
Machine learning algorithms cannot process raw text directly. Text must be converted into numerical values or vectors using Natural Language Processing (NLP) techniques:

1.  **Bag of Words (BoW):** Maps text to vectors containing raw word frequency counts.
2.  **TF-IDF:** Weights words based on local frequency and global rarity across documents.
3.  **Word2Vec:** Generates dense vector embeddings capturing deep semantic relationships.


## 3. Gaussian Naive Bayes
This variant is deployed when features contain continuous numerical values rather than categorical labels or discrete counts.

*   **Core Assumption:** Continuous features follow a **Normal (Gaussian) Distribution**. This is represented by a symmetric, bell-shaped curve centered around the mean ($\mu$).
*   **Classic Datasets:**
    *   *Iris Dataset:* Continuous measurements of plant petals and sepals.
    *   *Biometric Demographics:* Continuous features like `Age`, `Height`, and `Weight` to predict a discrete target output.

### Mathematical Approach
Instead of calculating probabilities via frequency counts, Gaussian Naive Bayes computes the mean ($\mu$) and variance ($\sigma^2$) of the continuous features for each class. It then utilizes the Gaussian Probability Density Function (PDF) to compute the conditional likelihood:

$$P(x_i \mid y) = \frac{1}{\sqrt{2\pi\sigma_y^2}} \exp\left(-\frac{(x_i - \mu_y)^2}{2\sigma_y^2}\right)$$


## 4. Variant Selection Summary Checklist

| **Multinomial NB** | Discrete frequency counts (e.g., Token counts) | Email spam filters, sentiment analysis |
| **Gaussian NB** | Continuous / Numerical features (e.g., Height, Weight, Age) | Iris flower classification, biometric prediction |


## 6. Mathematical Formulas for Naive Bayes


### 6.1 The General Bayes' Theorem Formula
Bayes' Theorem calculates the conditional probability (or "posterior probability") of an event $A$ occurring given that event $B$ has already occurred:

$$P(A \mid B) = \frac{P(B \mid A) \times P(A)}{P(B)}$$

Where:
*   $P(A \mid B)$ = **Posterior Probability**: The probability of hypothesis $A$ given the observed evidence $B$.
*   $P(B \mid A)$ = **Likelihood**: The probability of evidence $B$ given that hypothesis $A$ is true.
*   $P(A)$ = **Prior Probability**: The baseline probability of hypothesis $A$ before seeing any evidence.
*   $P(B)$ = **Marginal Likelihood (Evidence)**: The total probability of observing evidence $B$ across all outcomes.


### 6.2 The Naive Bayes Classifier Formula (For Feature Vectors)
When applied to machine learning, we predict a target class $y$ based on an input feature vector $\mathbf{X} = (x_1, x_2, \dots, x_n)$.

By applying the "naive" assumption—that all input features $x_i$ are conditionally independent of each other given the class $y$—the joint conditional probability in the numerator simplifies into a product of individual probabilities:

$$P(y \mid x_1, x_2, \dots, x_n) = \frac{P(y) \times \prod_{i=1}^{n} P(x_i \mid y)}{P(x_1, x_2, \dots, x_n)}$$

### The Optimization Problem (Argmax)
Because the denominator $P(x_1, x_2, \dots, x_n)$ is identical for every class being evaluated, it remains a constant and can be completely ignored when comparing probabilities.

To find the final predicted class ($\hat{y}$), we simply look for the class $y$ that maximizes the numerator:

$$\hat{y} = \arg\max_{y} \left( P(y) \times \prod_{i=1}^{n} P(x_i \mid y) \right)$$
